#### Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

#### Import Data

```python
df = pd.read_csv("BDC_changes_2.csv",sep= ",", header = None, names =["Datum","Uhrzeit","SECCI","Temperatur","Salinit√§t","NO2","NO3","NOx"],skiprows=2)
# Imports modified csv (wrong time/symbols manually changed)

df.drop(["Uhrzeit"], axis = 1, inplace =True)
# Drop Time

df.drop(["NOx"], axis = 1, inplace =True)
# Drop NOx

df["Datum"] = pd.to_datetime(df["Datum"], dayfirst=True)
# sets the Date to the datetime format

df["NO2"] = df["NO2"].astype((float))
# NO2 to float

df =df.set_index("Datum")
# set Date as Index

print(df.dtypes)
```

```python
df.plot(figsize=(20, 12))
```

# interpolate ("time")

```python
df_no_NaN_both_directions = df.interpolate(method="time",limit_direction="both")
# interpolation of the dataframe in both directions

df_no_NaN_both_directions.to_csv("df_cleaned_both_directions_interpolated.csv")
df_no_NaN_both_directions
```
